```python
import yfinance
import pyautogui
```

```python
ticker = input('Digite o código da ação: ') #aqui é o código da ação desejada
dados = yfinance.Ticker(ticker) #comando que importa as ações e coloca dentro da variavel dados
```

```python
dados.history()
```

```python
tabela = dados.history("6mo")
tabela
```

```python
fechamento = tabela.Close
fechamento
```

```python
fechamento.plot()
```

```python
maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]

print(maxima)
print(minima)
print(atual)
```

```python
# configurar uma pausa entre as ações do pyautogui
pyautogui.PAUSE = 6

# abrir uma nova aba
pyautogui.hotkey("ctrl", "t")

# copiar o endereço do gmail para o clipboard
pyperclip.copy("www.gmail.com")

# colar o endereço do gmail e dar um ENTER
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

# clicando no botão Escrever
pyautogui.click(x=39, y=144) #posição do mouse encontrado no código acima

# Preenchendo o destinatário
pyperclip.copy(destinatario)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab")

# Preenchendo o assunto
pyperclip.copy(assunto)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab")

# Preenchendo a mensagem
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

# Clicar no botão Enviar
pyautogui.click(x=1273, y=986) #posição do mouse encontrado no código acima

# fechar a aba do gmail
pyautogui.hotkey("ctrl", "f4")

# Imprimir mensagem de enviado com sucesso
print('E-mail enviado com sucesso!')
```
